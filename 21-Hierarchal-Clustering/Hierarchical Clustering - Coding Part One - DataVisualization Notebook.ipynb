```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/cluster_mpg.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_W_dummies = pd.get_dummies(df.drop('name', axis = 1))
```

```python
df_W_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_W_dummies)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data, columns = df_W_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize = (15, 8))
sns.heatmap(scaled_df)
```

```python
plt.figure(figsize = (15, 8))
sns.clustermap(scaled_df)
```

```python
sns.heatmap(scaled_df.corr())
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters = 4)
```

```python
scaled_df
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(data = df, x = 'mpg', y = 'weight', hue = cluster_labels)
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(data = df, x = 'mpg', y = 'horsepower', hue = cluster_labels,
               palette = 'viridis')
```

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold = 0)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
from scipy.cluster.hierarchy import dendrogram
```

```python
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize = (20, 10))
dendro = dendrogram(linkage_matrix, truncate_mode = 'lastp', p = 10)
```
